## 1. Combination Logic | flétturásir eða runurásir

#### Flétturásir

- Má skipta í **FLétturásir** eða **runurásir**
- Framkvæma aðgerðir sem lýst er með **Boolean jöfnum**
- **Útmerki(output) flétturása** er eingöngu háð núverandi **innmerkjum(input)** Enginn endurgjöf eða minniseiginleikar. 
- allar rökrásir sem við höfum skoðað hingað til eru **flétturásir**



- Flétturás hefur **n-marg innganga** og **m-marga útganga**
- **Innmerki(input-n)** geta verið **utanaðkomandi t.d frá öðrum flétturásum| *búnað eða skynjurum*
- **Útmerkin(output-m)** eru afurð flétturásarinnar

#### Runurásir

- Eru *flétturásir* sem hafa **minniseiginleika og endurgjöf** (e. feedback)
- **Útmerki(output)** ekki endilega háð núverandi innmerkjum(input) heldur **fyrri ástandi rásar**

## Samsettar grunneiningar

Til eru margar samsettar rásir sem koma oft fyrir í stafrænum rásum sem kalla mætti **grunneiningar**, þær helstu eru; 

- Samlagningarrásir (e. Adders +) 
- Frádráttarásir (e. Subtractors -)
- Samberar (e. Comparators >=<)
- Afkóðarar (e. Decoders)
- Fjölrásar (e. Multiplexers)

Allt þetta eru MSI = Medium Scale Integration

### Samlagningarrás 
Samlagning bita í stafrænum rásum er ein algengasta aðgerð sem við sjáum. Samlagningarás má skipta í 2x hluta. 
- Half-adder (HA)
- Full-adder (Fa)

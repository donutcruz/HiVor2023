a) 
- Í upphafi eru báðir vísarnir(index) með gildið 0
- Ef við setjum 1x stak(element) inn í biðröð(que) og tökum það strax út þá er biðröð(que) tóm og gildi beggja vísanna er 1. 
- Ef við höfum sett inn 2x stök og tekið þau út, þá eru báðir vísarnir(index) með gildið 2. o.s.frv

b)
- Vísirinn **head** er alltaf í hólfinu sem inniheldur fremsta stak(element)biðraðarinnar (que), nema þegar biðröðinn er tóm
- Vísirinn (index) **tail** er alltaf á hólfinu fyrir aftan síðasta stak biðraðarinnar (que)
- Þegar fylkið (array) er fullt þá er fremsti stað biðraðarinnar í þessu hólfi (biðröð vefst um fylgið *wraps*
- Vísirinn **Tail** er því í sama hólfinu og **head** og hafa **sama gildið** 
- Geta tekið gildi á bilinu 0 - capacity-1.
- Þetta þýðir að við sjáum ekki mun á fullri og tómri biðröð, út frá gildum þessara vísa 
- Við þurfum tilviksbreytuna(variable) **N** sem geymir fjölda staka í biðröðinni.
- Gildi hennar væri annaðhvort **0(ef tóm)** eða **capacity(ef full)**

**4. [Vaxtarhraði]**  Finnið vaxtarhraða á keyrslutíma sem fall af N fyrir forritsbútinn í 
æfingadæminu hér að ofan.  Sýnið vaxtarhraðann með slöngutáknum og rökstyðjið svar 
ykkar með útreikningum. 





- **Vaxtarhraði(e. growth rate)** er fall notaður til að lýsa því hvernig keyrslutími falls vex eftir því sem inntaksmagnið aukast.
- Algengt er slöngutákn **Big-O** = Lýsir yfirleitt versta mögulega keyrslutíma fallsins í verstu tilfelli

- **N** sé heilt veldi af 2 í forritsbútinum sem var lýst í fyrra svarinu. 
- Þá er síðasta gildið sem tekur i á sig gildið N, og heildarfjöldi skipta sem innsta skipunin keyrir er $2 + 4 + 8 + 16 + ... + 2N$.

Til að finna summu þessa raðar, getum við notað formúlu fyrir summu jafn stórra veldaröðu:

$$1 + 2 + 4 + 8 + ... + 2^n = 2^{n+1} - 1$$

Hér er $n$ fjöldi liða í veldaröðunni. Við getum nú notað þessa formúlu til að finna summu raðarinnar sem við erum að vinna með:

$$2 + 4 + 8 + 16 + ... + 2N = 2(1 + 2 + 4 + 8 + ... + N) = 2(2^{log_2 N + 1} - 1) = 2(2N - 1) = 4N - 2$$

Þar sem við lögðum saman veldaröðuna fyrir númer 0, 1, 2, ..., log(N), sem eru $log_2 N + 1$ liðir. Niðurstaðan er að heildarfjöldi skipta sem innsta skipunin keyrir er $4N - 2$.

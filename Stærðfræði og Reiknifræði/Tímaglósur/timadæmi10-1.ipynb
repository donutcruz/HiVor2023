## Æfing: Besta 3.stigs margliða

- Búið til 100. punkta gervigögn fyrir fallið y = x(x-2)(x-4) á bilinu [0,4]
- Finnið 3.stigs feril og teiknið hann og gögnin. 
- Hér þarf að nota lægra staðalfrávik á Delta en að ofan, t.d Delta = 0.1

```python
import numpy as np
import matplotlib as plt
import matplotlib.pyplot as plt


# Fylki með 100 jafnt bil milli 0,4 

x = np.linspace(0, 4, 100)

# reikna y-gildi

y = x * (x - 2 ) * (x - 4) 

# reikna staðalfrávik 0.1

y_std = np.std(y)
print('Staðalfrávik =', y_std)

# finna 3.stigs feril með Numpy.polyfit

feril = np.polyfit(x, y, 3)
p = np.poly1d(feril)

# stærð á Ásanna 

plt.figure(figsize= (6, 6))
# teikna gögnin og 3.stigs ferilinn

plt.plot(x,y, '', label = 'gögnin',)
plt.plot(x, p(x), '', label = '3.stigs jafna')
plt.legend()
plt.show()
```
